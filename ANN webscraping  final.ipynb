```python
#importing libraries
import requests     # get url requests
from bs4 import BeautifulSoup #beautiful soup for reading html content
import re #Regular Expression
import numpy as np 
import pandas as pd
```

```python
# Make a request to the website
url = 'https://www.aan.com/msa/public/events/index/38'
response = requests.get(url)
```

```python
# Parse the HTML content
soup = BeautifulSoup(response.content, 'lxml')
```

```python
#initializing variables
internal_links = []
#Event_name = []
#Date = []
#Time = []
#Topic = []
#Event_type = []
First_tenlinks = []
```

```python
#Finding All the Internal links of event pages
for link in soup.find_all(
'a', href=re.compile('^(/MSA/Public/Events/Details)((?!:).)*$')):
    internal_links.append(link)
```

```python
#using only first 20 links 
First_tenlinks = internal_links[:30]
```

```python
#First_tenlinks
```

```python
#creating list variable for storing internal links with http
Links_with_http = []
```

```python
# Loop over each internal link and scrape data
for link in First_tenlinks:
    href = link.get('href')
    if href and not href.startswith(('http', 'https')):#checking url Contains https or http
        # Prepend the base URL to the href
        href = 'https://www.aan.com' + href
        #print(href)
        Links_with_http.append(href)
        
        
```

```python
len(Links_with_http)
```

```python
#Events_details = []
#Events_details.clear()
#Event_name.clear()
```

```python
Events_details = []
Event_name = []
Date = []
Time = []
Topic = []
Event_type = []
presentation_title = []
Faculty = []
Facultydisclosure = []

#Loop for storing html content of  event details inside the inside links
for link in Links_with_http:
    html = requests.get(link)
    soup = BeautifulSoup(html.content,'lxml')
    Events_details =soup.find('div',class_='singlestep-form-fields')
    component_containers = soup.find_all('div',class_='component-container')
    if len(component_containers) >= 3:
        Event_timings = component_containers[2]
        Faculty_dis = component_containers[3]
        # rest of your code for this case
    else:
        continue
        # code to handle the case when there are not enough elements in the list
        #Event_timings = soup.find_all('div',class_='component-container')[2]
        #Faculty_dis = soup.find_all('div',class_='component-container')[3]
    try:
        Event_name.append(Events_details.find('h2').text.strip())
    except:
        Event_name.append('NA')
  #  Event_name.append(Events_details.find('h2').text.strip())
    try:
        Date.append( Events_details.find('fieldset',id="Date").text.strip())
    except:
        Date.append('NA')
   #Date.append( Events_details.find('fieldset',id="Date").text.strip())
    try:
        Time.append( Events_details.find('fieldset',id="Time").text.strip())
    except:
        Time.append('NA')
#    Time.append( Events_details.find('fieldset',id="Time").text.strip())
    try:
        Event_type.append( Events_details.find('fieldset',id="Type").text.strip())
    except:
        Event_type.append("NA")
#     Event_type.append( Events_details.find('fieldset',id="Type").text.strip())
    try:
         Topic.append( Events_details.find('fieldset',id="Topics").text.strip())
    except:
        Topic.append('NA')
#    Topic.append( Events_details.find('fieldset',id="Topics").text.strip())
    try:
        presentation_title.append(Event_timings.find_all('td')[2].text.strip())
    except:
        presentation_title.append('NA')
#        presentation_title.append(Event_timings.find_all('td')[2].text.strip())
    try:
        Faculty.append(Faculty_dis.find_all('td')[0].text.strip())
    except:
        Faculty.append('NA')
#        Faculty.append(Faculty_dis.find_all('td')[0].text.strip())

    try:
        Facultydisclosure.append(Faculty_dis.find_all('td')[1].text.strip())
    except:
        Faculty.append('NA')
#        Facultydisclosure.append(Faculty_dis.find_all('td')[1].text.strip())
```
